```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a scatterplot colored by species using FacetGrid
g = sns.FacetGrid(iris, hue='species', height=5)
g.map(plt.scatter, 'sepal_length', 'sepal_width')
g.add_legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a boxplot to visualize PetalLengthCm by species
sns.boxplot(x='species', y='petal_length', data=iris)

# Set labels for axes
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Set the plot title
plt.title('Boxplot of Petal Length by Species')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a boxplot with individual points using stripplot
ax = sns.boxplot(data=iris, x='species', y='petal_length')
ax = sns.stripplot(data=iris, x='species', y='petal_length', jitter=True, edgecolor='green')

# Set labels for axes
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Set the plot title
plt.title('Boxplot with Stripplot of Petal Length by Species')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a violin plot to visualize the distribution of PetalLengthCm by species
sns.violinplot(x='species', y='petal_length', data=iris, height=6)

# Set labels for axes
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Set the plot title
plt.title('Violin Plot of Petal Length by Species')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a FacetGrid with kdeplots for PetalLengthCm by species
sns.FacetGrid(iris, hue="species", height=6) \
   .map(sns.kdeplot, "petal_length") \
   .add_legend()

# Set labels for axes
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')

# Set the plot title
plt.title('Kernel Density Estimation of Petal Length by Species')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a pairplot colored by species
sns.pairplot(iris, hue='species', height=3)

# Show the plot
plt.show()
```

Another useful seaborn plot is the pairplot, which shows the bivariate relation between each pair of features

From the pairplot, we'll see that the Iris-setosa species is separataed from the other two across all feature combinations
